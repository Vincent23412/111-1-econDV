```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv("C:/Users/asd11/OneDrive/桌面/covid19_tw_specimen.csv")
```

```python
dates = []
for i in range(len(data)) :
    year,month,_ = data['通報日'][i].split("/")
    if int(month) < 10 :
        date = int(str(year) + "0" + str(month) + "01")
    else :
        date = int(str(year) + str(month) + "01")
    dates.append(date)
data['月份'] = dates
```

```python
data.head()
```

```python
data = data.groupby(["月份"]).agg({'Total':sum})
```

```python
rate = [0]
for i in range(1,len(data)) :
    a = (data.values[i-1] + data.values[i]) / data.values[i-1]
    rate.append(a[0])
data['成長率'] = rate
```
