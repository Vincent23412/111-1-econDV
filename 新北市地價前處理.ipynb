```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/asd11/Downloads/新北市110年公告土地現值.csv")
```

```python
data.head()
```

```python
average_price = data.groupby(by='district').agg({'official_value':np.mean}).sort_values(by='official_value',ascending=False)
average_price.columns = ['Average_price']
average_price[:5]
```

```python
land_sum = data.groupby(by='district').agg({'official_value':len}).sort_values(by='official_value',ascending=False)
land_sum.columns = ['Total_land']
land_sum[10:]
```

```python
average_price[:5].join(land_sum,how='inner')
```

```python
fliter = (data['district'] == '永和區') | (data['district'] == '板橋區') | (data['district'] == '三重區') | (data['district'] == '中和區') | (data['district'] == '蘆洲區')
new_data = data[fliter]
```

```python
new_data.to_csv('新北市地價現值.csv')
```
